```python
import h5py
import torch
```

```python
path = "data/PCOD2.h5"
# Read H5py train,val, and test sets
with h5py.File(path, 'r') as f:
    pointclouds = []
    # print(type(f))
    for synsetid in ["1111"]: # synsetids are groups
        for split in ('train', 'val', 'test'): # Splits are datasets

          print(type(f[synsetid]))
          print(type(f[synsetid][split]))
          print(type(f[synsetid][split][...]))
          pointclouds.append(torch.from_numpy(f[synsetid][split][...]))
```
